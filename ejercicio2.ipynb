```python
import spacy

# Cargar el modelo de lenguaje español
nlp = spacy.load("es_core_news_sm")

# Leer el contenido del archivo
with open("discurso.txt", "r", encoding="utf-8") as file:
    texto = file.read()

# Procesar el texto con spaCy
doc = nlp(texto)

# Extraer y presentar las oraciones
oraciones = list(doc.sents)
for i, oracion in enumerate(oraciones):
    print(f"Oración {i + 1}: {oracion.text.strip()}")

# Presentar el número de oraciones
print(f"Número total de oraciones: {len(oraciones)}")
```
